## Reason Drop Missing Value
The decision to drop data based on a high percentage of missing values depends on several factors, such as the percentage of missing values, the importance of the missing data, the reason for the missing data, and the overall size of the data set.

There is no specific percentage of missing values that is universally considered "high" and warrants dropping the data. However, as a general rule of thumb, if a variable has more than 50% missing values, it may be appropriate to consider dropping that variable from the analysis Our data missing value percentage is 52% and it is more than 50. 
so we need to drop that.

### Check for outliers by either plotting box plots or IQR For all numerical variables

```python
# Extract the numerical variables
numerical_vars = df.select_dtypes(include=['float64', 'int64']).columns
  
for var in numerical_vars:
    sns.boxplot(x=df[var])
    plt.title(var)
    plt.show()

# Calculate the IQR for each numerical variable and count the number of outliers
for var in numerical_vars:
    Q1 = df[var].quantile(0.25)
    Q3 = df[var].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5*IQR
    upper_bound = Q3 + 1.5*IQR
    outliers = df[(df[var] < lower_bound) | (df[var] > upper_bound)][var]
    print("Number of outliers for", var, ":", len(outliers))
```

### Write a query that returns:

#### a) What percentage of female customers own a house? 

```python
%%sql
SELECT Gender, round((COUNT(*) * 1.0 /
                      (SELECT COUNT(*)
                       FROM Customers
                       WHERE Gender = 'F'))* 100,2) AS Percentage
FROM PurHistory
WHERE Gender = 'F' AND Cust_Prop = 'H'
GROUP BY Gender;
```

#### b) How much is the average salary of these customers?

```python
%%sql
SELECT round(AVG(Cust_Ann_Income),2) as AverageSalary
FROM PurHistory
```
